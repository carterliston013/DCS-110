```python
df.isna().sum()
```

This code uses the 'isna' command showing which values in columns do not have an actual value. This establishes the 'sum' or amount of n/a values in each column.

```python
df.isna().mean()*100
```

This code uses the same isna function but uses the 'mean' command and a multiplication by 100 to return a percentage of the total values in each column that have a null value.

```python
df_clean = df.dropna(subset=['has_diabetes']).copy()
```

This code gets rid of or 'drops' all of the n/a values from the 'has_diabetes' column, creating a copy without these data points allowing us to begin cleaning the dataframe, establishing this copy as 'df_clean'.

```python
df_clean.isna().mean()*100
```

This repeats the percentage of n/a command to verify that our previous code was successful.

```python
df_clean.shape
```

This code once again shows us the 'shape' of our new copied dataframe, returning the rows and columns that remain in our data.

```python
mean_bp = df_clean['systolic_bp'].mean()
```

This code establishes the variable 'mean_bp' as the mean of values in the column 'systolic_bp' from the df_clean data frame.

```python
print(mean_bp)
```

This uses the print command to return the mean value of systolic bp as established in the previous command.

```python
df_clean['systolic_bp'] = df_clean['systolic_bp'].fillna(mean_bp)
```

This code uses the 'fillna' command to fill the n/a values with the mean of all values in the systolic_bp column.

```python
df_clean['systolic_bp'].isna().sum()
```

This command shows the n/a values that remain in the 'systolic_bp' column, returning that there are '0' values left as n/a after our previous commands to fill them.

**Transform Variables**

```python
df['bmi_category'] = pd.cut(df['bmi'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

This establishes the dataframe's bmi_category as subsets inputting the df 'bmi' using the bins 0, 18.5, 25, 30, and 100 as parameters and labeling them 'Underweight', 'Normal', 'Overweight', 'Obese'.

```python
df['bmi_category'].head()
```

This code once again shows the first 5 rows in the established bmi category of the dataframe.

```python
df['bmi_category'].value_counts()
```

This code counts the amount of values that are included in each subset of the bmi category.

**Categorical Variable Transformation**

```python
df['education'].head()
```

This code shows all of the unique values that are included in the education column, using the 'unique' command.

```python
education_map={
    1.0: 'Less than 9th grade',
    2.0: '9-11th grade',
    3.0: 'High school graduate/GED or equivalent',
    4.0: 'Some college',
    5.0: 'College graduate or above',
    9.0: 'Do not know',
    'nan': 'Missing'
}
```

This uses the education_map to display the meanings of each of the unique values preseneted from the previous command.

```python
df['education_label']=df['education'].map(education_map)
```

This command maps the values in the education map to the labels and values in the education column, allowing for more descriptive analytics.

```python
df['education_label'].value_counts()
```

This code counts the number of values for each of the different categories that have been mapped from the previous command in the education column.
