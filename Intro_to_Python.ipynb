# **Functions**

```python
temp = (4 * 9/5) + 32
print(temp)
```

```python
temp = (-6 * 9/5) + 32
print(temp)
```

```python
def celsius_to_fahrenheit (celsius):
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit
```

```python
celsius_to_fahrenheit (100)
```

Formatted String Literals

```python
print(f'The temperature now is {celsius_to_fahrenheit(472)}F')
```

Loops

```python
My_Temperature_List = [0, 11, 28, 42, 69]
```

```python
celsius_to_fahrenheit(0)
celsius_to_fahrenheit(10)
celsius_to_fahrenheit(11)
```

```python
for celsius in My_Temperature_List:
  print (celsius_to_fahrenheit(celsius))
```

# BMI Function

```python
def BMI_Calc (height, weight):
  BMI = (weight/ height**2) * 703
  return BMI
```

```python
BMI_Calc (73, 180)
```

```python
My_BMI_List = [(73, 180), (60, 250), (82,100)]
```

```python
for (height, weight)in My_BMI_List:
  print(BMI_Calc(height, weight))
```
