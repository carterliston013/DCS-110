```python
def celsius_to_fahrenheit (celsius):
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit
```

This defines a Celsius to Fahrenheit calculation function with Celsius as the main variable and Fahrenheit as the returned solution value.

```python
celsius_to_fahrenheit (100)
```

This uses the established Celsius to Fahrenheit function to return the Fahrenheit solution (212) from the Celsius variable (100).

Formatted String Literals

```python
print(f'The temperature now is {celsius_to_fahrenheit(472)}F')
```

This prints a sentence, while including a functional result by beginning with the f function allowing a calculation to be done in part with the sentence.

This command inputs the values of My_Temperature_List into the Celsius to Fahrenheit function as the variable Celsius to return the Fahrenheit values associated with the values in the list.

# BMI Function

```python
def BMI_Calc (height, weight):
  BMI = (weight/ height**2) * 703
  return BMI
```

This is a function defined for calculating BMI using weight in lbs and height in inches.

```python
BMI_Calc (73, 180)
```

This uses the values 73 (for height in inches) and 180 (weight in lbs) to return the solution of the BMI for these values.

```python
My_BMI_List = [(73, 180), (60, 250), (82,100)]
```

This establishes values for the (height, weight) input into a list called My_BMI_List.

```python
for (height, weight)in My_BMI_List:
  print(BMI_Calc(height, weight))
```

This command uses the values from My_BMI_List to input them into the BMI_Calc as (height, weight) and return their BMI values respectively.

```python
BMI=18
if BMI<20:
  print('Underweight')
else:
  print('Normal')
```

This establishes ranges for text returns according to BMI values, BMI<20 being Underweight and all other values as normal.
